## Load the data
1. Read the IGN data file into a dataframe named `ign_scores`.  
2. Use the `"Platform"` column to label the rows.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the IGN scores data
IGN= "https://raw.githubusercontent.com/csbfx/advpy122-data/master/ign_scores.csv"
ign_scores = pd.read_csv(IGN)

# Set the 'Platform' column as the index
ign_scores = ign_scores.set_index("Platform")
```

## Problem 1 - (1 point)
Use the dataframe ign_scores and write one line of code to determine the highest score received by any kind of games on the Wii platform?

```python
# Find the highest score across all genres for the Wii platform
highest_score_wii = ign_scores.loc["Wii"].max()
print(f"The highest score on the Wii platform is: {highest_score_wii}")


```

## Problem 2 - (1 point)
Use the dataframe `ign_scores` to determine which genre has the lowest score for the `PlayStation Vita` platform.

```python
# Find the genre with the lowest score for the PlayStation Vita platform
lowest_genre_ps_vita = ign_scores.loc["PlayStation Vita"].idxmin()
lowest_score_ps_vita = ign_scores.loc["PlayStation Vita"].min()

print(f"The genre with the lowest score for the PlayStation Vita platform is: {lowest_genre_ps_vita}")
print(f"The lowest score is: {lowest_score_ps_vita}")

```

## Problem 3 -  (2 points)
Your instructor's favorite video game has been Mario Kart Wii, a racing game released for the Wii platform in 2008. And, IGN agrees with her that it is a great game -- their rating for this game is a whopping 8.9! Inspired by the success of this game, your instructor is considering creating your very own racing game for the Wii platform. Perform the following analyses to help her determine which platform she should focus on.

1. Create a bar chart that shows the score for *Racing* games, for each platform. Your chart should have one bar for each platform. Provide a meaningful title to the plot.

2. Based on the bar chart, do you expect a racing game for the **Wii** platform to receive a high rating?  If not, use the pandas to find out from the dataframe `ign_scores` which gaming platform is the best for racing game?

```python
# (1 point)
# Set the width and height of the figure
import matplotlib.pyplot as plt

# Extract the Racing game scores for each platform
racing_scores = ign_scores["Racing"]

# Create the bar chart
plt.figure(figsize=(12, 6))  # Set the width and height of the figure
racing_scores.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title("Racing Game Scores Across Platforms", fontsize=14)
plt.xlabel("Platform", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
## (1 point)
## Use ign_scores to determine which gaming platform is the best
## for racing game.

# Find the platform with the highest average Racing game score
best_racing_platform = racing_scores.idxmax()
best_racing_score = racing_scores.max()

print(f"The best platform for racing games is {best_racing_platform} with a score of {best_racing_score}.")
```

```python
## Create a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Create a heatmap of IGN scores by genre and platform
plt.figure(figsize=(15, 10))
sns.heatmap(ign_scores, annot=True, fmt=".2f", cmap="YlGnBu", linewidths=0.5)
plt.title("Heatmap of IGN Scores by Genre and Platform", fontsize=16)
plt.xlabel("Genres", fontsize=12)
plt.ylabel("Platforms", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()

```

## Problem 5 - (4 points)
Use the Pokemon dataset to create a clustermap with color. First, filter the dataframe to only keep data with `Type 1` equals to following values: `Water`, `Normal`, `Grass`, `Bug` and `Psychic`. Annotate the dendrogram using different colors for these five different `Type 1` values. Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

```python
# Problem 5
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Pokemon dataset
pokemon_url = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv"
pokemon_data = pd.read_csv(pokemon_url)

# Filter the dataset for Type 1 values: Water, Normal, Grass, Bug, Psychic
filtered_data = pokemon_data[pokemon_data['Type 1'].isin(['Water', 'Normal', 'Grass', 'Bug', 'Psychic'])]

# Set 'Name' as the index
filtered_data = filtered_data.set_index('Name')

# Drop non-numeric columns to keep only the features for clustering
numeric_data = filtered_data.select_dtypes(include=['float64', 'int64'])

# Create a clustermap
plt.figure(figsize=(12, 8))
clustermap = sns.clustermap(
    numeric_data,
    metric="euclidean",
    method="average",
    cmap="viridis",
    figsize=(12, 8),
    row_colors=filtered_data['Type 1'].map({
        'Water': 'blue',
        'Normal': 'grey',
        'Grass': 'green',
        'Bug': 'orange',
        'Psychic': 'purple'
    }),
    annot=False
)

# Add a legend for Type 1 colors
from matplotlib.patches import Patch
legend_patches = [
    Patch(color='blue', label='Water'),
    Patch(color='grey', label='Normal'),
    Patch(color='green', label='Grass'),
    Patch(color='orange', label='Bug'),
    Patch(color='purple', label='Psychic')
]
plt.legend(handles=legend_patches, loc='upper right', title="Type 1", bbox_to_anchor=(1.2, 1))

plt.title("Clustermap of Selected Pokémon Types", fontsize=16)
plt.show()
```
